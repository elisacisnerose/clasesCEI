## Entrada y salida de datos

La manera en que tenemos los humanos de interactuar con las interfaces y los programas es mediante, en muchos casos, la entrada manual de datos y las salidas de texto en la interfaz gráfica. Para ello, Python permite hacerlo de manera sencilla.

### Salida de datos

La salida de datos la entendemos como aquellos datos que el programa proporciona al exterior. En Python lo haremos mediante la función **print()**.

```python
print("Hola")
saludo = "Buenas tardes"
print(saludo)
```

La función **print()** admite varios argumentos seguidos.
Para que los interprete correctamente, estos deben separarse por comas. En cambio, cuando se trate de dos cadenas seguidas, se puede no escribir comas entre ellas, aunque Python no las separará con un espacio.

```python
print("Hola", "Adiós")
print ("Hola" "Adiós")
```

La función **print()** permite incluir variables o expresiones como argumentos, lo que nos permite combinar texto y variables.

```python
nombre = "Pedro"
edad = 31
print("Me llamo", nombre, "y tengo", edad, "años.")
```

```python
semanas = 5
print("En", semanas, "semanas hay", 7*semanas, "días")
```

Además de esta manera, existen las **cadenas f**, la cual nos permite combinar texto y variables de manera más amigable y fluida.
Para poder utilizarlas, añadiremos una **f** antes de comenzar el texto de salida de datos y encapsularemos las variables entre corchetes **{ }**.

```python
nombre = "Pedro"
edad = 31
print(f"Me llamo {nombre} y tengo {edad} años.")
```

```python
semanas = 5
print(f"En {semanas} semanas hay {7*semanas} días")
```

```python
print("""
Hola, hoy es miércoles.
Mañana es jueves.
""")
```

### Entrada de datos

En este caso, la entrada de un programa son los datos que llegan desde el exterior, y que en este caso los introduciremos mediante el teclado.
Para ello, tenemos la función integrada **input()** que nos permitirá introducir por teclado todo aquello que deseemos.

```python
nombre = input("¿Cómo se llama? ")
print(f"Encantado de conocerle, {nombre}")
```

Lo más destacable de esta función es que todo lo que se introduce por teclado, Python lo almacena como una cadena de texto. Esto implica que si el valor solicitado es un valor numérico (entero o decimal) hemos de convertirlo al formato que queramos para poder realizar el tratamiento del dato adecuado.

```python
edad = int(input("Dígame su edad: "))
#edad = input("Dígame su edad: ")
print(f"Dentro de 5 años tendrás {edad+5}")
```

```python
altura = float(input("Dígame su altura (en metros) ") )
print(f"Usted mide {altura*100} centímetros")
```

Si queremos personalizar el mensaje de entrada, la función **input()** también permite cadenas f, por lo que podemos añadir variables en el mensaje de entrada.

```python
nombre = input("¿Cómo se llama? ")
apellido = input(f"¿Y cómo es su primer apellido, {nombre}? ")
print(f"Encantado de conocerle, {nombre} {apellido}")
```
