##1. Introducción

Todo el mundo sabe lo que es un objeto: Una cosa tangible que podemos sentir y manipular. Los primeros objetos con los que interactuamos suelen ser juguetes. Los bloques de madera, las figuras de plástico, rompecabezas.. son los primeros objetos con los que interactuamos. Rápidamente aprendemos que ciertos objetos hacen ciertas cosas: las campanas suenan, los botones se pulsan y las palancas se accionan.
<br>
<br>
La definición de “objeto” en desarrollo de software no es muy diferente, con la diferencia de que los objetos de software no son cosas tangibles que se puedan coger o manipular físicamente, son modelos de algo que puede hacer ciertas cosas.
Formalmente, un objeto es una colección de datos y comportamientos asociados. “Orientación a objetos”, no es más que un estilo de desarrollo software.
<br>
<br>
El Diseño Orientado a Objetos es el proceso de convertir los requisitos funcionales de un software en una especificación de implementación.
El diseñador debe nombrar los objetos, definir los comportamientos y especificar cómo los objetos pueden activar ciertos comportamientos. La etapa de diseño trata sobre cómo deben hacerse las cosas. Al igual que antes de empezar a construir un edificio son necesarios unos planos, antes de empezar a construir un software, es necesaria una etapa de diseño.
<br>
<br>
Por desgracia, en el mundo real muchas veces esta etapa es la gran olvidada.

##2. UML (Unified Modeling Language)

UML es un estándar muy común en desarrollo de software para visualizar y documentar sistemas y aplicaciones software.
<br>
Proporciona una notación gráfica que permite a los miembros del equipo de desarrollo representar de manera clara y precisa la estructura, comportamiento y relaciones que se producen dentro de una aplicación de software.
<br>
<br>
La sintaxis de un diagrama UML suele ser bastante obvia, intuitiva, fácil de aprender y facilitar la comunicación entre programadores, aunque a veces este lenguaje sea muy menospreciado por los programadores, ya que algunos piensan que hacer un diseño previo en UML es una pérdida de tiempo. Esto es un gran error, ya que al utilizar UML los desarrolladores pueden visualizar y comprender mejor la arquitectura y el comportamiento de un sistema, lo que ayuda a reducir errores de diseño y facilitar la colaboración entre miembros del equipo.
<br>
<br>
La ventaja de UML es que la mayoría de las cosas son opcionales. Sólo necesitamos especificar la información que sea necesaria para nuestro caso particular.
<br>
<br>
En resumen, UML es una herramienta esencial en procesos de desarrollo software para modelar y documentar aplicaciones de software de manera clara y coherente.

##3. Clases y Objetos

Como se comentó en la introducción, un objeto es una colección de datos con comportamientos asociados.
<br>
Las Clases son una especie de plantillas o moldes que definen la estructura y comportamiento de un conjunto de objetos. Estas clases describen cómo deben ser los objetos, especificando sus atributos y métodos.
<br>
<br>
Por otro lado, los objetos son las instancias concretas creadas a partir de estas clases, representan entidades reales que tienen valores específicos para sus atributos y pueden ejecutar los métodos definidos en su respectiva clase.
<br>
<br>
Como resumen, podemos decir que las clases representan una estructura y los objetos representan una realidad.
<br>
<br>
<br>
**Para entender cada parte de este módulo utilizaremos este ejemplo:**
<br>

<font color='blue'>Tenemos una clase genérica de Televisión.

* Esta clase tiene una serie de características como pueden ser: la marca, el modelo, el color, la resolución o el tamaño de la pantalla y una serie de comportamientos o acciones asociados a ella como pueden ser: el hecho de encenderla, apagarla o modificar el volumen.
* Cada televisión que se construya siguiendo la plantilla anteriormente definida, será una instancia de la clase Televisión, también llamada “objeto”.

##4. Especificación de atributos y métodos

Como se indicó en el anterior punto, los objetos son instancias de clases que pueden asociarse entre sí.

Una instancia es un objeto específico de una clase con su propio conjunto de datos y comportamientos.
<br>
<br>
* Los datos suelen representar las características individuales de un determinado objeto. Una clase puede definir conjuntos específicos de características que comparten todos los objetos de esa clase, a estas características las llamamos atributos.

> <font color='blue'>En nuestro ejemplo diseñando la clase Televisión, podríamos tener como atributos: la marca, el modelo, el color, la resolución o el tamaño de pantalla.</font>

image.png

* Los comportamientos son acciones que pueden ocurrir en un objeto, a estos comportamientos les llamamos métodos. Los métodos son funciones que tienen acceso a todos los datos asociados con ese objeto (atributos), por tanto, al igual que las funciones, los métodos también pueden aceptar parámetros y devolver cosas. Los parámetros de un método son una serie de entradas que deben pasarse al método para que éste pueda realizar una tarea y devolver el resultado de esa tarea.

> <font color='blue'>En nuestro ejemplo, nuestra clase Televisión podría tener métodos para: encender, apagar o métodos para modificar el volumen.</font>

##5. Interfaz pública y encapsulación

La interfaz es una colección de atributos y métodos que se pueden utilizar con un objeto. El principal objetivo del Diseño Orientado a Objetos es determinar cuál va a ser la interfaz pública de ese objeto con la que podrán interactuar los usuarios u otros objetos.

<font color='blue'>En nuestro caso, la interfaz pública de nuestro objeto televisión, sería el mando a distancia. Cada botón del mando representa un método que puede ser llamado en el objeto televisión.</font>
<br>
<br>
image.png

Aparte de la interfaz pública, también existen una serie de atributos y métodos privados a los que el usuario no debe poder acceder.

<font color='blue'>Siguiendo el ejemplo de la televisión, cada vez que nosotros accionamos una acción a través del mando, ocurren una serie de cosas transparentes para el usuario. Por ejemplo, al pulsar el botón para subir el volumen, el mando emite una señal, llega al receptor de la televisión, y la televisión internamente regula la señal electrónica que le manda a los altavoces.
<br>
<br>
Como usuarios, siempre que queremos regular el volumen, lo hacemos a través de la interfaz (el mando), en ningún momento se nos ocurre desmontar la televisión y regular la señal electrónica de los altavoces a mano (de hacerlo así, anularíamos la garantía).
</font>

Este proceso de ocultar la implementación o los detalles funcionales de un objeto se denomina Abstracción. Por tanto, la abstracción es el proceso de encapsular información.

<font color='blue'>Siguiendo el ejemplo anterior, nuestra clase Televisión dispondría de métodos privados para que el mando a distancia mande la señal o para que el circuito electrónico de los altavoces se ajuste con el volumen adecuado. También dispondría de un atributo privado que haga referencia al volumen actual de la televisión, que sólo puede ser modificado a través de la interfaz pública.
</font>
<br>
<br>
image.png

**Recomendaciones**
<br>
* A la hora de diseñar, debemos tener en cuenta la importancia de la interfaz pública ya que es muy difícil cambiarla en el futuro. Por eso debemos diseñar siempre la interfaz pública de la manera más sencilla y fácil de usar posible.
<br>
<br>
* Debemos asegurarnos que los métodos y atributos tengan nombres coherentes, para ello podemos seguir estas pautas:
> * Los objetos deben representar sustantivos.
> * Los métodos deben representar verbos.
> * Los atributos deben representar adjetivos.
<br>
* No debemos modelar objetos con acciones o atributos que pensamos que podrían ser útiles en el futuro. Debemos modelar exactamente las tareas que la aplicación necesita realizar, y el diseño irá evolucionando de manera natural a un nivel de abstracción apropiado.


##6. Patrones de diseño

Hasta ahora, hemos aprendido a diseñar sistemas como un grupo de objetos que interactúan, donde cada interacción implica ver los objetos a un nivel apropiado de abstracción, pero aún no sabemos cómo crear estos niveles de abstracción.
<br>
<br>
Los Patrones de Diseño son soluciones probadas y estandarizadas para aplicar a problemas comunes de diseño de software. Proporcionan un enfoque estructurado y eficiente para resolver desafíos de diseño, mejorando la calidad del código y la flexibilidad del sistema.
<br>
<br>
Dos de los principios fundamentales en el Diseño Orientado a Objetos son la Composición y la Herencia.
<br>
<br>
<font color='blue'>Siguiendo el ejemplo anterior, supongamos que queremos diseñar una clase RemoteControl que pueda interactuar con un objeto de tipo Television.</font>


### 6.1 Composición

El principio de composición se basa en la idea de que las clases deberían lograr la reutilización de código y la flexibilidad a través de la composición de objetos.
<br>
<br>
La composición permite que una clase contenga instancias de otras clases como atributos y, de esta manera, pueda utilizar sus funcionalidades sin heredar todas sus características, esto fomenta un diseño más flexible y modular.
<br>
<br>
* Para indicar la composición entre clases, se unen con línea continua la Clase de contenedor con la Clase de contenido.
* La punta de la flecha tiene forma de rombo y se coloca en el diagrama de la Clase contenedora.

<font color='blue'>Supongamos que queremos diseñar una clase llamada Speaker con métodos para subir y bajar el volumen, de manera que estos altavoces sean una clase independiente. Esto permitirá que también se puedan reutilizar en otras clases que implementen altavoces. Esto es un ejemplo de Composición, tendríamos el siguiente resultado: La clase Television contiene un atributo que es una instancia de la clase Speaker.</font>
<br>
<br>

### 6.2 Herencia

La herencia es otro principio fundamental en la programación orientada a objetos y se refiere a la capacidad de una clase de heredar propiedades y métodos de una clase padre.
<br>
<br>
La herencia se utiliza cuando existe una relación de “es un” entre clases. Sin embargo, es importante usarla con moderación, ya que puede llevar a una jerarquía de clases compleja y acoplada.
<br>
<br>
* Para indicar la herencia entre clases, se unen con línea continua la Clase padre con la Clase hija.
* El extremo de la línea tiene forma de flecha y apunta al diagrama de la Clase padre.

<font color='blue'>Supongamos que ahora queremos diseñar la clase SmartTv, esta clase va a tener las mismas características que la clase Television, pero además va a tener un nuevo atributo que hace referencia al sistema operativo, y un nuevo comportamiento que hace referencia a abrir la aplicación Netflix.

Para resolver este problema debemos aplicar la Herencia: Crearemos una clase SmartTv (clase hija) que herede las características de la clase Television (clase padre) y luego extienda o modifique su comportamiento según sea necesario.</font>
